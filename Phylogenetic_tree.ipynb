```python
pip install ete3
import pandas as pd
from ete3 import NCBITaxa
ncbi = NCBITaxa()
ncbi.update_taxonomy_database()
```

```python
#Extract taxon ids
taxonids = pd.read_csv('C:/Users/tneth/OneDrive/Documents/MSc_AppBio/MSc_project/metadata/taxonids.csv', delimiter=",", encoding='ISO-8859-1', dtype=str)
df = pd.DataFrame(taxonids)

df['ids'] = df['taxon_id'].str.replace('\xa0', '', regex=False).astype(int)
taxon_list = df['ids'].tolist()
print(taxon_list)
```

```python
#Extract topology of taxons
tree = ncbi.get_topology(taxon_list)

#Display leaf name
for leaf in tree.iter_leaves():
    leafname=int(leaf.name)
    name = ncbi.get_taxid_translator([leafname])[leafname]
    leaf.name = name

#Display node name
for i, node in enumerate(tree.traverse()):
    if not node.is_leaf():
        nodename=int(node.name)
        name = ncbi.get_taxid_translator([nodename])[nodename]
        node.name = name
        print(node.name)
```

```python
#Save phylogenetic tree as a newick file
tree.write(format=1, outfile='tree_7.nwk')
```
